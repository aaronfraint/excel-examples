# Vectorized calculations

... are a major benefit of using `pandas`. You can write a single formula that then gets applied to every row in your table.

The starting data has a column called `annual_salary` but each row is actually a quarter.

In the example below I create a new column named `quarterly_salary` and set the value as `annual_salary` divided by 4

```python
import pandas as pd

df = pd.read_csv("../data/employee_salaries.csv")
```

```python
# print the dataframe to see the starting state

df.head()
```

```python
# Create a new column and set the value as a formula that refers to another existing column

df["quarterly_salary"] = df["annual_salary"] / 4
```

```python
# print the dataframe again, showing the newly added column

df.head()
```
